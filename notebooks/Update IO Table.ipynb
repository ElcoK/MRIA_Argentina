```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
from ras_method import ras_method
import subprocess
data_path = os.path.join('..','Data')
```

```python
sectors = [chr(i) for i in range(ord('A'),ord('P')+1)]
```

```python
IO_2004 = pd.read_csv(os.path.join(data_path,'economic_IO_tables','input','Argentina_IO_2004.csv'),index_col=[0],header=[0],sep=';')*1000
```

```python
ValueA_series = pd.read_excel(os.path.join(data_path,'economic_IO_tables','input','sh_VBP_VAB_06_19.xls'),sheet_name='Cuadro 4',skiprows=3,index_col=[0])/1e3
```

```python
Total_Prod = pd.read_excel(os.path.join(data_path,'economic_IO_tables','input','sh_VBP_VAB_06_19.xls'),sheet_name='Cuadro 2',skiprows=3,index_col=[0])/1e3
```

```python
IO_2004*1e6
```

```python
FD = np.array(IO_2004['FD'][:16]*np.array((Total_Prod[2004]))/np.array(IO_2004.sum(1)[:16]))
Exports = np.array(IO_2004['EXP'][:16][:16]*np.array((Total_Prod[2004]))/np.array(IO_2004.sum(1)[:16]))
Imports = np.array(IO_2004.loc['Imports'][:16]*np.array((Total_Prod[2004]))/np.array(IO_2004.sum(1)[:16]))
ValueA = np.array(ValueA_series[2004])
InterDem =  np.array(Total_Prod[2004]-Imports-ValueA)

# convert to numpy matrix
X0 = IO_2004.values[:,:]

# get sum of T
u = np.array(list(Total_Prod[2004])+[sum(ValueA),sum(Imports)])
v = np.array(list(Total_Prod[2004])+[x*((sum(ValueA)+sum(Imports))/(sum(FD)+sum(Exports))) for x in [sum(FD),sum(Exports)]])
v[v < 0] = 0
# and only keep T

# apply RAS method to rebalance the table
new_IO = ras_method(X0,u,v,1e-5,print_out=True)
```

```python
NEW_IO = pd.DataFrame(new_IO,index=sectors+['ValueA','Imports'],columns=sectors+['FD','EXP'])
```

```python
for year in [int(x) for x in np.linspace(2004, 2015, 12)]:
    FD = np.array(NEW_IO['FD'][:16]*np.array((Total_Prod[year]))/np.array(NEW_IO.sum(1)[:16]))
    Exports = np.array(NEW_IO['EXP'][:16][:16]*np.array((Total_Prod[year]))/np.array(NEW_IO.sum(1)[:16]))
    Imports = np.array(NEW_IO.loc['Imports'][:16]*np.array((Total_Prod[year]))/np.array(NEW_IO.sum(1)[:16]))
    ValueA = np.array(ValueA_series[year])
    InterDem =  np.array(Total_Prod[year]-Imports-ValueA)

    # convert to numpy matrix
    X0 = NEW_IO.values[:,:]

    # get sum of T
    u = np.array(list(Total_Prod[year])+[sum(ValueA),sum(Imports)])
    v = np.array(list(Total_Prod[year])+[x*((sum(ValueA)+sum(Imports))/(sum(FD)+sum(Exports))) for x in [sum(FD),sum(Exports)]])
    v[v < 0] = 0
    # and only keep T

    # apply RAS method to rebalance the table
    new_IO = ras_method(X0,u,v,1e-5,print_out=True)
    
    NEW_IO = pd.DataFrame(new_IO,index=sectors+['ValueA','Imports'],columns=sectors+['FD','EXP'])
```

```python
NEW_IO.to_csv(os.path.join(data_path,'economic_IO_tables','input','Argentina_IO_2015.csv'))
```
